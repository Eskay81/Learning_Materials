```python
print("Shape is ",Car.shape)
#print(Car.size)
```

```python
CarRight = Car + [2,0]
```

```python
plt.plot(Car[:,0],Car[:,1],CarRight[:,0],CarRight[:,1])
plt.grid()
plt.title("Car moved along X axis 2 units")
```

```python
CarUp = Car - [0,2]

plt.plot(Car[:,0],Car[:,1],CarUp[:,0],CarUp[:,1])
plt.grid()
plt.title("Car moved along Y axis 2 units")
```

```python
CarUp = Car - [2,2]

plt.plot(Car[:,0],Car[:,1],CarUp[:,0],CarUp[:,1])
plt.grid()
plt.title("Car moved along X and Y axis 2 units")
```
